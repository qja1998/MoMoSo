```python
import requests
import json

from dotenv import load_dotenv
import os

# load .env
load_dotenv()

# Hugging Face Inference API Endpoint 및 토큰 설정
HF_ENDPOINT = "https://api-inference.huggingface.co/models/DeepSeek-R1"
HF_API_KEY = os.environ.get('HF_API_KEY')
HEADERS = {"Authorization": f"Bearer {HF_API_KEY}"}

def generate_novel(prompt: str, max_length: int = 500):
    """
    Hugging Face API를 호출하여 소설을 생성하는 함수
    """
    payload = {
        "inputs": prompt,
        "parameters": {
            "max_length": max_length,
            "temperature": 0.8,
            "top_p": 0.95,
            "do_sample": True
        }
    }
    
    response = requests.post(HF_ENDPOINT, headers=HEADERS, data=json.dumps(payload))
    
    if response.status_code == 200:
        return response.json()
    else:
        return {"error": "Failed to generate text", "status_code": response.status_code}
```
