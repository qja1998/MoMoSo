```python
import requests
import dotenv
import os

dotenv.load_dotenv()

JUPYTER_URL = os.environ["JUPYTER_URL"]
TOKEN = os.environ["TOKEN"]

headers = {
    "Authorization": f"token {TOKEN}"
}

# Jupyter 서버 정보 확인
response = requests.get(f"{JUPYTER_URL}", headers=headers)

if response.status_code == 200:
    print("Jupyter API 접근 성공:", response.json())
else:
    print("API 접근 실패:", response.status_code, response.text)
```

```python
data = {
    "code": "result = 5 * 10"
}

response = requests.post(JUPYTER_URL + "/run", json=data)
print(response.json())
```
